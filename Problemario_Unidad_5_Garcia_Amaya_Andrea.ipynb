##Problema 1:
Se hace un estudio sobre la efectividad de tres marcas de atomizador para matar moscas. Para ello, cada producto se aplica a un grupo de 100 moscas, y se cuenta el número de moscas muertas expresado en porcentajes. Se hicieron seis réplicas, pero en días diferentes; por ello, se sospecha que puede haber algún efecto importante debido a esta fuente de variación. Los datos obtenidos se muestran a continuación:

| Marca de atomizador | Réplica 1 | Réplica 2 | Réplica 3 | Réplica 4 | Réplica 5 | Réplica 6 |
|----------------------|-----------|-----------|-----------|-----------|-----------|-----------|
| 1                   | 72        | 65        | 67        | 75        | 62        | 73        |
| 2                   | 55        | 59        | 68        | 70        | 53        | 50        |
| 3                   | 64        | 74        | 61        | 58        | 51        | 69        |

**Preguntas:**<br>
a) Suponiendo un DBCA, formule las hipótesis adecuadas y el modelo estadístico.  
b) ¿Existe diferencia entre la efectividad promedio de los atomizadores?  
c) ¿Hay algún atomizador mejor? Argumente su respuesta.  
d) ¿Hay diferencias significativas en los resultados de diferentes días en que se realizó el experimento? Argumente su respuesta.  
e) Verifique los supuestos de normalidad y de igual varianza entre las marcas.


## Respuetas
Diseño de Bloques Completos al Azar (DBCA)

**a)**
Hipotesis para marcas:
H0: No hay diferencia en la efectividad promedio entre las tres marcas de atomizador.
H1: Al menos una marca tiene diferencia  en la efectividad promedio.

Hipostesis para dias:
H0: No hay diferencia en los resultados debida a los días de réplica.
H1: Al menos un día produce resultados diferentes.

Modelo estadístico: Se utilizo ANOVA.

**b)** Segun ANOVA con un valor - p para marcas de 0.1028 no se rechaza la hipotesis nula por lo tanto no hay diferencia significativa entre la efectividad promedio de los atomizadores.

**c)** No por que que no hay diferencias significativas entre las marcas entonces no se puede identificar un atomizador que sea mejor con base en estos datos.

**d)** El valor - p para réplicas (días) es 0.420718 entonces no hay diferencia significativa en los días en que se realizó el experimento.

**e)** No se rechaza normalidad de residuos y no se rechaza homocedasticidad entonces si se cumplen los supuestos.

```python
import pandas as pd
from io import StringIO

data = '''
marca,replica,moscas
1,1,72
1,2,65
1,3,67
1,4,75
1,5,62
1,6,73
2,1,55
2,2,59
2,3,68
2,4,70
2,5,53
2,6,50
3,1,64
3,2,74
3,3,61
3,4,58
3,5,51
3,6,69
'''

df = pd.read_csv(StringIO(data))
df
#Andrea Garcia Amaya
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Realizar el modelo ANOVA de dos vías
modelo_lineal = ols(f'moscas ~ C(marca) + C(replica)', data=df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal, typ=2)
tabla_anova
#Andrea Garcia Amaya
```

```python
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

residuos = modelo_lineal.resid

# Prueba de Shapiro-Wilk
stat, valor_p_sh = shapiro(residuos)
print(f"valor-p (Shapiro) = {valor_p_sh}")

# Visualización: Q-Q plot
sm.qqplot(residuos, line='s')
plt.title("Q-Q plot de residuos")
plt.show()

# Histograma
sns.histplot(residuos, kde=True)
plt.title("Histograma de residuos")
plt.show()
#Andrea Garcia Amaya
```

```python
from scipy.stats import levene

# Supón que tienes un DataFrame df con columnas 'respuesta' y 'grupo'
grupo1 = df[df['marca'] == 1]['moscas']
grupo2 = df[df['marca'] == 2]['moscas']
grupo3 = df[df['marca'] == 3]['moscas']


stat, p = levene(grupo1, grupo2, grupo3)
print(f"p-valor de Levene: {p}")
#Andrea Garcia Amaya
```

```python
eta_squared = tabla_anova['sum_sq']['C(marca)'] / tabla_anova['sum_sq'].sum()
print(f"η² (tamaño de efecto): {eta_squared:.3f}")
#Andrea Garcia Amaya
```

**a)** Diseño en Bloques Completos al Azar (DBCA)

H1: Al menos un detergente difiere en blancura.

Hipotesis para lavadoras:
H0: No hay diferencia en la blancura debido a las lavadoras.

**c)** Detergentes: El ANOVA arrojo un valor - p de 0.000363 entonces se rechaza H0 y si hay diferencias significativas en el rendimiento (blancura) de los detergentes.

```python
import pandas as pd
from io import StringIO

data = '''
detergente,lavadora,blancura
A,1,45
A,2,43
A,3,51
B,1,47
B,2,44
B,3,52
C,1,50
C,2,49
C,3,57
D,1,42
D,2,37
D,3,49
'''

df = pd.read_csv(StringIO(data))
df
#Andrea Garcia Amaya
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Realizar el modelo ANOVA de dos vías
modelo_lineal = ols(f'blancura ~ C(detergente) + C(lavadora)', data=df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal, typ=2)
tabla_anova
#Andrea Garcia Amaya
```
